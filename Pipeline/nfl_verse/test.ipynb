```python
team_df = pd.read_csv('pfr.csv')

for week in team_df['Week'].unique():
    pass_y = team_df[team_df['Week'] == week]['PassY']
    if not pass_y.empty:
        print(week, " : ", pass_y.iloc[0])
```

```python
for week in df['week'].unique():
    print(week)

    yards = 0
    attempts = 0
    plays = []

    for _, play in df[df['week'] == week].iterrows():
        if (
            play['posteam'] == team and # offensive play
            play['season_type'] == 'REG' and # reg szn plays
            play['rush_attempt'] == 1 and # is a pass attempt
            play['two_point_conv_result'] == None # not a two point try
        ):
            attempts += 1
            plays.append(play)
            
            yards += play['yards_gained']
    pass_y = team_df[team_df['Week'] == week]['RushY']
    if not pass_y.empty:
        pass_y = pass_y.iloc[0]

    if pass_y != yards:
        print(pass_y, "!=", yards)     
    print(attempts , ":", yards)
```

```python
team = "DAL"
for week in df['week'].unique():
    print(week)
    yards = 0
    attempts = 0
    plays = []

    for _, play in df[df['week'] == week].iterrows():
        if (
            play['posteam'] == team and # offensive play
            play['season_type'] == 'REG' and # reg szn plays
            play['pass_attempt'] == 1 and # is a pass attempt
            play['two_point_conv_result'] == None # not a two point try
        ):
            if play['sack'] == 1:
                yards += play['yards_gained']
                continue
            attempts += 1
            plays.append(play)
            
            yards += play['yards_gained']
    pass_y = team_df[team_df['Week'] == week]['PassY']
    if not pass_y.empty:
        pass_y = pass_y.iloc[0]

    if pass_y != yards:
        print(pass_y, "!=", yards)     

```

```python
print(len(plays))

df_plays = pd.DataFrame(plays)
df_plays.to_csv("week_12_verse.csv", index=False)
```
